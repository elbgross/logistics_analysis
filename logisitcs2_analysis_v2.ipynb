```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

print("Path to dataset files:", path)
```

```python
import os

os.listdir(path)
```

```python
import os
import pandas as pd

orders = pd.read_csv(os.path.join(path, 'olist_orders_dataset.csv'), parse_dates=[
        "order_purchase_timestamp",
        "order_estimated_delivery_date",
        "order_delivered_customer_date"
    ])
customers = pd.read_csv(os.path.join(path, 'olist_customers_dataset.csv'))
sellers = pd.read_csv(os.path.join(path, 'olist_sellers_dataset.csv'))
items = pd.read_csv(os.path.join(path, 'olist_order_items_dataset.csv'))

#parse_dates is used to indicate that certain columns should be parsed as datetime objects.
```

```python
items

# Aggregate items data to get total price and quantity per order
```

```python
items_agg = (
    items
    .groupby("order_id")
    .agg(
        total_price=("price", "sum"),
        total_freight=("freight_value", "sum"),
        n_items=("order_item_id", "count"),
        seller_id=("seller_id", "first")
    )
    .reset_index()
)

items_agg
```

```python
#Create a dataframe merging orders, customers, sellers, and aggregated items data. Useful for EDA and also for our ml model.


df = (
    orders
    .merge(items_agg, on="order_id", how="left")
    .merge(customers[["customer_id", "customer_city", "customer_state"]],
           on="customer_id", how="left")
    .merge(sellers[["seller_id", "seller_city", "seller_state"]],
           on="seller_id", how="left")
)

df.to_csv("ecommerce_data_enriched.csv", index=False)
```

```python
#We only use columns with data that is not available at the time of order placement to define the target but NOT for the features.
# Create delay target variable


df["delay"] = (
    df["order_delivered_customer_date"] >
    df["order_estimated_delivery_date"]
).astype(int)

```

```python
df["purchase_weekday"] = df["order_purchase_timestamp"].dt.weekday
df["purchase_hour"] = df["order_purchase_timestamp"].dt.hour
df["purchase_month"] = df["order_purchase_timestamp"].dt.month
df["purchase_weekofyear"] = df["order_purchase_timestamp"].dt.isocalendar().week
df["purchase_weekend"]= df["purchase_weekday"].isin([5,6]).astype(int)
```

```python
df["route"] = df["seller_state"] + "_" + df["customer_state"]
```

```python
features = [
    "purchase_weekday",
    "purchase_hour",
    "purchase_month",
    "purchase_weekofyear",
    "purchase_weekend",
    "total_price",
    "total_freight",
    "n_items",
    "customer_city",
    "customer_state",
    "seller_city",
    "seller_state",
    "route"
]

X = df[features]
y = df["delay"]
```

```python
#Split the data into training and test sets based on order purchase date. Because date has impact on delays.

split_date = "2018-01-01"

train_mask = df["order_purchase_timestamp"] < split_date
test_mask  = df["order_purchase_timestamp"] >= split_date

X_train, X_test = X[train_mask], X[test_mask]
y_train, y_test = y[train_mask], y[test_mask]
```

```python
# Impute missing values in numerical columns with median. Logisitic Regression cannot handle missing values.

num_cols = X_train.select_dtypes(include=["number"]).columns
medians = X_train[num_cols].median()

X_train[num_cols] = X_train[num_cols].fillna(medians)
X_test[num_cols]  = X_test[num_cols].fillna(medians)
```

```python
X_train = pd.get_dummies(X_train, drop_first=True)
X_test  = pd.get_dummies(X_test, drop_first=True)
```

```python
# Align columns in case some categories are missing in test or train set

X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)
```

```python
#Imbalanced target only the 7,48 orders are delayed. I use OverSampling because the dataset already have too many noise.
# Noise in a dataset refers in this case that same inputs can have different outputs. I wouldn´t risk it using SMOTE.


from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)

X_train_res, y_train_res = ros.fit_resample(
    X_train,
    y_train
)

print("Before oversampling:")
print(y_train.value_counts(normalize=True))

print("\nAfter oversampling:")
print(y_train_res.value_counts(normalize=True))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import roc_auc_score

rf = RandomForestClassifier(
    random_state=42,
    n_jobs=1
)


rf_param_dist = {
    "n_estimators": [200, 400, 800],
    "max_depth": [None, 8, 12, 16, 24],
    "min_samples_split": [2, 5, 10, 20],
    "min_samples_leaf": [1, 2, 5, 10],
    "max_features": ["sqrt", "log2", 0.3, 0.5, 0.8],
    "bootstrap": [True],
    "class_weight": [None, "balanced"]}

rf_search = RandomizedSearchCV(
    rf,
    rf_param_dist,
    n_iter=12,
    scoring="roc_auc",
    cv=2,
    random_state=42,
    n_jobs=-1,   
    verbose=1
)
```

```python
rf_search.fit(X_train_res, y_train_res)

```

```python
print("Best params:", rf_search.best_params_)
```

```python
best_rf = rf_search.best_estimator_
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import roc_auc_score

y_prob = best_rf.predict_proba(X_test)[:, 1]




roc_auc = roc_auc_score(y_test, y_prob)
print("ROC-AUC:", roc_auc)
```

Given the high dimensionality introduced by one-hot encoding and the relatively weak signal, a regularized Logistic Regression model was evaluated as a strong linear baseline.

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    solver="liblinear"
)
```

```python
log_reg.fit(X_train_res, y_train_res)
```

```python
y_prob_log = log_reg.predict_proba(X_test)[:, 1]

roc_auc = roc_auc_score(y_test, y_prob_log)
print("ROC-AUC:", roc_auc)
```

### **Conclusions**

In this notebook, I trained two ML models to predict delivery delays in orders: Random Forest and Logistic Regression. The evaluation was based on ROC-AUC, where Random Forest achieved 0.55 and Logistic Regression 0.61.

Overall, both models show **limited predictive power**, which suggests that delays cannot be accurately predicted as a strict yes/no outcome for each individual order. However, the results indicate that it is still possible to estimate the **risk of delay**: some orders can be classified as more likely to be delayed than others.

Therefore, this problem is better approached as a **risk prediction task**, useful for prioritizing monitoring and preventive actions rather than expecting perfect delay prediction per order.
